```python
names = ['Neg', 'CMV', 'Pos']
amj = [amj_neg, amj_cmv, amj_seb]
ncols = len(amj)
stride = 25

figure(figsize=(12,4))
dim = dims[0]
for i, data in enumerate(amj):
    subplot(1,ncols,i+1)
    x = data[::stride, dim[0]]
    y = data[::stride, dim[1]]
    z = bilinear_interpolate(x, y)
    scatter(x, y, s=1, c=z, edgecolors='none')
    axis([-10000, 90000, -10000, 100000])
    title(names[i], fontsize=14)
    xticks([])
    yticks([])
    fill([48000,48000,70000,70000] ,[98000,50000,50000,98000], ec='red', fc='none')
tight_layout()
```
