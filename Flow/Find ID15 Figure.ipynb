```python
import os, sys, glob
import fcm
import utils_cc as utils
from matplotlib.path import Path
from matplotlib.nxutils import points_inside_poly
```

```python
def _idx(x, max_events):
    idx = np.arange(len(x))
    np.random.shuffle(idx)
    return idx[:max_events]

def subsample(xs, max_events):
    ys = []
    for x in xs:
        idx = _idx(x, max_events)
        y = x[idx, :]
        ys.append(y)
    return ys
```

```python
cols = [0,1,4,2,3]
base_dir = '/data/cliburn/CIMT3'
x = fcm.loadFCS(os.path.join(base_dir, 'data/ID15_donor2/CMV.fcs'))
y = x[:, cols]
```

```python
ys = array(subsample([y], 100000))[0]
```

```python
boundaries = {'cd8': (1, 3), 'fsc': (-0.75, 1.25), 'tet': (2.5, 10), 'ssc': (-1.5, 0.5), 'cd3': (0, 2)}
mus = y.mean(0)
sds = y.std(0)
fsc_limits = mus[0] + boundaries['fsc'][0]*sds[0], mus[0] + boundaries['fsc'][1]*sds[0]
ssc_limits = mus[1] + boundaries['ssc'][0]*sds[1], mus[1] + boundaries['ssc'][1]*sds[1]
cd3_limits = mus[2] + boundaries['cd3'][0]*sds[2], mus[2] + boundaries['cd3'][1]*sds[2]
cd8_limits = mus[3] + boundaries['cd8'][0]*sds[3], mus[3] + boundaries['cd8'][1]*sds[3]
tet_limits = mus[4] + boundaries['tet'][0]*sds[4], mus[4] + boundaries['tet'][1]*sds[4]
box0 = array([[fsc_limits[0], fsc_limits[0], fsc_limits[1], fsc_limits[1]], [ssc_limits[1], ssc_limits[0], ssc_limits[0], ssc_limits[1]]])
box1 = array([[cd3_limits[0], cd3_limits[0], cd3_limits[1], cd3_limits[1]], [cd8_limits[1], cd8_limits[0], cd8_limits[0], cd8_limits[1]]])
box2 = array([[cd8_limits[0], cd8_limits[0], cd8_limits[1], cd8_limits[1]], [tet_limits[1], tet_limits[0], tet_limits[0], tet_limits[1]]])
boxes = [box0, box1, box2]
```
