```python
base_dir = '/data/cliburn/EQAPOL_normal/data'
```

```python
xss = np.load(os.path.join(base_dir, 'xss.npy'))
```

```python
box0 = array([[50000,50000,250000,250000] ,[100000,5000,5000,100000]])
box1 = array([[48000,48000,200000,150000] ,[98000,50000,50000,98000]])
box2 = array([[25000,25000,70000,70000] ,[5000,25000,25000,5000]])
boxes = [box0, box1, box2]
```

```python
xs = xss[0]
dims = [(0,3), (1,2), (5,4)]
ncols = len(dims)
nrows = 1

hs = ['FSC-A', 'FSC-H', 'FSC-W', 'SSC-A', 'Amine', 'CD3']
figure(figsize=(12,4))

for k, dim in enumerate(dims):
    subplot(nrows, ncols, k + 1)
    x = xs[:10000, dim[0]]
    y = xs[:10000, dim[1]]
    z = bilinear_interpolate(x, y)
    scatter(x, y, s=1, c=z, edgecolors='none')
 
    xticks([])
    yticks([])
    if k==0 or k==1:
        axis([0,265000,0,265000])
    else:
        axis([-10000, 100000, -10000, 100000])
    xlabel(hs[dim[0]], fontsize=16)
    ylabel(hs[dim[1]], fontsize=16)
      
    fill(boxes[k][0], boxes[k][1], ec='red', fc='none')
tight_layout()
```
