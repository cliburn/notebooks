Extract 8 peak bead data for normalization of EP3 data
Data is only available on lejeune and discordia.

```python
import os, sys, glob
import fcm
from fcm.graphics import bilinear_interpolate
import fcm.statistics as stats
import utils
from matplotlib.path import Path
import reflowrestclient.utils as rest
```

```python
base_dir = '/home/cliburn/Dropbox/Inbox/Data/EP3/'

fs = sorted(glob.glob(os.path.join(base_dir, '*bead*fcs')))
fs
```

```python
long_names = []
for fidx in range(len(fs)):
    path, filename = os.path.split(fs[fidx])
    name, ext = os.path.splitext(filename)
    site = name[:3]
    compensation = os.path.join(base_dir, 'CMw%s' % site)
    sidx, spill = fcm.load_compensate_matrix(compensation)
    xs = fcm.loadFCS(os.path.join(path, filename), spill=spill, sidx=sidx, transform='logicle')
    ln = [n for n in xs.long_names if not (('SC-' in n) or ('Time' in n))]
    long_names.append([site] + ln)
    
for item in array(long_names):
    print item[0], [i.ljust(18) for i in item[1:]], len(item)
```

```python
channels = array(xs.channels)
ys = xs.copy()

if channels[0] == 'Time':
    tmp = channels[0]
    channels[:-1] = channels[1:]
    channels[-1] = tmp
    tmp1 = ys[:, 0]
    ys[:, :-1] = ys[:, 1:]
    ys[:, -1] = tmp1
    
zip(range(100), channels)
```
