```python
import fcm
import fcm.statistics as stats
import os, sys, glob
import utils
```

Preprocess data by extracting specific columns for arcsinh transform
----

```python
try:
    fs = np.load('data/fs.npy')
    hs = np.load('data/hs.npy')
    xs = np.load('data/xs.npy')
except:
    a=0
    b=1/5.0
    c=0
    cols = np.array([12, 14, 17, 22, 23, 26, 27, 29, 
                    30, 32, 33, 35, 36, 37, 39, 40, 41,]) - 1
    xs = []
    hs = []
    fs = []
    for f in glob.glob('/data/cliburn/Nima_CyTOF/*'):
        path, filename = os.path.split(f)
        fs.append(filename)
        x = fcm.loadFCS(f, transform=None, auto_comp=False)
        hs.append(np.array(x.channels)[cols])
        x = x[:, cols]
        x = np.arcsinh(a + b*x) + c
        xs.append(x)
        
    np.save('data/fs.npy', fs)
    np.save('data/hs.npy', hs)
    np.save('data/xs.npy', xs)
```

Define HDPGMM model
----

```python
nclusts = 32
niter = 1000
burnin = 10000
tag = 'k%d_i%d_b%d' % (nclusts, niter, burnin)

m = stats.HDPMixtureModel(nclusts=nclusts, niter=niter, burnin=burnin)
m.ident = True
```

Fit model to data and suummarize - use average of posterior and identify modes
----

```python
try:
    cs = utils.unpickle('data/%s_cs.pic.gz' % tag)
except:
    rss = m.fit(xs, verbose=True)
    rs = rss.average()
    cs = rs.make_modal()
    utils.pickle(rss, 'data/%s_rss.pic.gz' % tag)
    utils.pickle(rs, 'data/%s_rs.pic.gz' % tag)
    utils.pickle(cs, 'data/%s_cs.pic.gz' % tag)
```

Classify original events with modal cluster labels
----

```python
try:
    zs = np.load('data/%s_zs.npy' % tag)
except:
    zs = [c.classify(x) for c, x in zip(cs, xs)]
    np.save('data/%s_zs.npy' % tag, zs)
```
