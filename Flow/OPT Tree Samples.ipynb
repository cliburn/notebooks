```python
import fcm
import fcm.statistics as stats
import os, sys, glob
from fcm.graphics import bilinear_interpolate
import utils
```

```python
sidx, spill = fcm.load_compensate_matrix('/data/cliburn/EQAPOL_normal/data/CompMatrixDenny06Nov09')
xcos = fcm.loadFCS('/data/cliburn/EQAPOL_normal/data/AMJ_5L_Costim.fcs', sidx=sidx, spill=spill, transform='logicle')
xcmv = fcm.loadFCS('/data/cliburn/EQAPOL_normal/data/AMJ_5L_CMV pp65.fcs', sidx=sidx, spill=spill, transform='logicle')
xseb = fcm.loadFCS('/data/cliburn/EQAPOL_normal/data/AMJ_5L_SEB.fcs', sidx=sidx, spill=spill, transform='logicle')
```

```python
dim = [12, 15]
ycos = xcos[np.random.choice(range(len(xcos)), 10000), :]
ycos = ycos[:, dim]
ycmv = xcmv[np.random.choice(range(len(xcmv)), 10000), :]
ycmv = ycmv[:, dim]
yseb = xseb[np.random.choice(range(len(xseb)), 10000), :]
yseb = yseb[:, dim]
ys = [ycos, ycmv, yseb]
```

```python
centroids = [r.mus for r in rs]
zs = [r.classify(y) for (r, y) in zip(rs, ys)]

ratio_threshold = 0.75

targets = []
print '%2s: %4s %4s %4s %4s %4s' % ('ID', 'ncos', 'ncmv', 'nseb', 'diff', 'ratio')
for i in range(len(centroids[0])):
    ncos = np.sum(zs[0]==i)
    ncmv = np.sum(zs[1]==i)
    nseb = np.sum(zs[2]==i)
    if nseb:
        ratio = 1.0*nseb/(1+ncos)
        diff = nseb - ncos
        if ((ncos <= ncmv and ncmv <= nseb) or (ncos >= ncmv and ncmv >= nseb)) and ((ratio < ratio_threshold) or (ratio > 1.0/ratio_threshold)):
            print '%2d: %4d %4d %4d %4d %.2f' % (i, ncos, ncmv, nseb, diff, ratio)
            targets.append(i)

figure(figsize=(12,4))
labels = ['Neg', 'CMV', 'SEB']
for i, y in enumerate(ys):
    subplot(1,3,i+1)
    scatter(y[:, 0], y[:,1], s=1, c=bilinear_interpolate(y[:,0], y[:,1]), edgecolors='none')
    scatter(y[zs[i]==4, 0], y[zs[i]==4, 1], s=10, c='red')
    xticks([])
    yticks([])
    axis([-10000,110000,-10000,110000])
    text(0, 100000, labels[i], fontsize=16, color='red')
    for k, m in enumerate(centroids[i]):
        if k in targets:
            text(m[0], m[1], str(k), fontsize=16, va='center', ha='center', bbox=dict(facecolor='yellow', alpha=0.5))
tight_layout()
```

```python
figure(figsize=(12, 4))
labels = ['Neg', 'CMV', 'SEB']

for i, _ys in enumerate(ys):
    z = zs[i]
    idx = np.zeros(len(ys), 'bool')
    for t in targets:
        idx |= z==t
    
    
    subplot(1,3,i+1)
    scatter(_ys[:, 0], _ys[:,1], s=1, c='grey', alpha=0.3, edgecolors='none')
    scatter(_ys[idx, 0], _ys[idx,1], s=5, c=z[idx], alpha=1.0, edgecolors='none')
    xticks([])
    yticks([])
    axis([-10000,110000,-10000,110000])
    text(0, 100000, labels[i], fontsize=16, color='red')
tight_layout()
```
