```python
import matplotlib
matplotlib.rcParams['font.size'] = 16
warnings.simplefilter("ignore", RankWarning);
```

```python
def f(x, eps=0):
    return 0.5*x**2 + 4 + eps * random.normal(size=len(x))

figure(figsize=(15,6))
xs = arange(-3, 3.1, 0.5)
xtrain = xs[1:-1:2]
xtest = xs[::2]

print "Trainig set", xtrain
print "Test set", xtest

xi = arange(-3, 3.1, 0.1)
y = f(xtrain)
y1 = f(xtrain, 1)

subplot(2,5,1)
plot(xtrain, y, 'o')
p = polyfit(xtrain, y, 2)
yhat = polyval(p, xi)
plot(xi, yhat, 'r-')
xlim([x[0]-0.1, x[-1]+0.1])
title('True (no noise)')

ps = []
for i, degree in enumerate(range(1, 9)):
    subplot(2,5, i+2)
    plot(xtrain, y1, 'o')
    p = polyfit(xtrain, y1, degree)
    ps.append(p)
    yhat = polyval(p, xi)
    plot(xi, yhat, '-')
    xlim([xs[0]-0.1, xs[-1]+0.1])
    title('Degree = %d' % degree)
tight_layout()
```

```python
errors = []
y2 = f(xtest)
for i, p in enumerate(ps):
    yhat = polyval(p, xtest)
    errors.append(sum((y2 - yhat)**2))
    
plot(range(1, len(errors)+1), errors, '-o')
xlim([0.5, len(errors)+0.5])
xlabel('Degree of Polynomial')
ylabel('Error on test set');
```
