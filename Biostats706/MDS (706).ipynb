```python
from sklearn import manifold
from sklearn.metrics import euclidean_distances
import matplotlib
matplotlib.rcParams['font.size'] = 16
```

```python
n = 100
ndim = 10
xs = np.concatenate([np.random.multivariate_normal(zeros(ndim), eye(ndim), n),
                     np.random.multivariate_normal(zeros(ndim)+[0,0,10,0,10,0,0,0,0,0], eye(ndim), n)])
xs.shape
```

```python
figure(figsize=(8,4))
subplot(121)
scatter(xs[:100,0], xs[:100,1])
scatter(xs[100:,0], xs[100:,1], c='red')
xlabel('X0')
ylabel('X1')
title('No sap in 8/10 dims')
subplot(122)
scatter(xs[:100,2], xs[:100,4])
scatter(xs[100:,2], xs[100:,4], c='red', s=30)
xlabel('X2')
ylabel('X4')
title('Sep in 2/10 dims')
tight_layout()
```

```python
seed = np.random.RandomState(seed=3)
similarities = euclidean_distances(xs)
mds = manifold.MDS(n_components=2, max_iter=3000, eps=1e-9, random_state=seed,
                   dissimilarity="precomputed", n_jobs=1)
pos = mds.fit(similarities).embedding_
```

```python
scatter(pos[:100, 0], pos[:100, 1])
scatter(pos[100:, 0], pos[100:, 1], c='red')
pos.shape
```
