```python
mu = array([0.0, 0.0])
sigma = array([[1.0, 0.9],[0.9, 1.0]])
sigma
```

```python
xs = np.random.multivariate_normal(mu, sigma, 100)
```

```python
axis('equal')
scatter(xs[:, 0], xs[:, 1]);
xlabel('X1')
ylabel('X2')
```

```python
sxs = (xs - xs.mean(0))/xs.std(0)
```

```python
mu_hat = sxs.mean(0)
sigma_hat = cov(sxs.T)
```

```python
# using SVD of data matrix
eigenvectors, eigenvalues, V = np.linalg.svd(sxs.T, full_matrices=False)
print eigenvectors
print eigenvalues
```

```python
# Using eigenvalues and eigenvectors of covariance matrix
eigenvalues, eigenvectors = eig(sigma_hat)
print eigenvectors
print eigenvalues
```

```python
def plot_pc(name, start, end):
    annotate(name, xy=end, xycoords='data', xytext=start, textcoords='data', 
             arrowprops=dict(facecolor='red', arrowstyle="->"), fontsize=20)
```

```python
axis('equal')
scatter(xs[:, 0], xs[:, 1], facecolors='none', edgecolors='blue')
for i, v in enumerate(eigenvectors):
    plot_pc('', mu_hat, mu_hat + dot(sigma_hat, v))
xlabel('X')
ylabel('Y')
```

```python
axis('equal')
projected_data = dot(sxs, eigenvectors)
scatter(projected_data[:, 0], projected_data[:, 1], facecolors='none', edgecolors='blue')
xlabel('PC1')
ylabel('PC2')
```
