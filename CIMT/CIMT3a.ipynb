```python
import os, sys, glob
import fcm
from fcm.graphics import bilinear_interpolate
```

```python
def pickle(fname, obj):
    import cPickle, gzip
    cPickle.dump(obj=obj, file=gzip.open(fname, "wb", compresslevel=3), protocol=2)

def unpickle(fname):
    import cPickle, gzip
    return cPickle.load(gzip.open(fname, "rb"))
```

```python
fs = sorted(glob.glob('/data/cliburn/CIMT2/data/ID01/*fcs'))
```

```python
base_dir = '/data/cliburn/CIMT3/data'
col_dict = unpickle('/home/cchan/hg/analysis/cimt3/col_dict.pic.gz')

try:
    xss = np.load(os.path.join(base_dir, '%s_viability_xss.npy' % lab))
    channels = np.load(os.path.join(base_dir, '%s_viability_channels.npy' % lab))
    names = np.load(os.path.join(base_dir, '%s_viability_names.npy' % lab))
except:    
    names = []
    xss = []
    channels = []
    for f in fs:
        path, filename = os.path.split(f)
        name, ext = os.path.splitext(filename)
        lab, donor, stim = name.split('_')
        
        print donor
        
        key = '../data/ID01_%s/%s.fcs' % (donor, stim)
        if key in col_dict:
           pmap = col_dict[key]
        else:
           pmap = col_dict[key.replace('donor', 'Donor')]
        
        cols = [int(pmap[s]) for s in ['fsc-h', 'ssc-h', 'cd3', 'cd8', 'tet', 'stained']]
        print key, cols
        
        xs = fcm.loadFCS(f, auto_compensate=True, transform='logicle')
        hs = [xs.long_names[i] for i in cols]
        ys = xs[:, cols]
        
        names.append([lab, donor, stim])    
        xss.append(ys)
        channels.append(hs)
        
    np.save(os.path.join(base_dir, '%s_viability_xss.npy' % lab), xss)
    np.save(os.path.join(base_dir, '%s_viability_channels.npy' % lab), channels)
    np.save(os.path.join(base_dir, '%s_viability_names.npy' % lab), names)
```

```python
import fcm.statistics as stats

nclusts = 320
burnin = 10000
niter = 100
ident = '%s_k%d_b%d_i%d' % (lab, nclusts, burnin, niter)
n = 100000

def sample(xs, n):
    idx = np.arange(len(xs))
    np.random.shuffle(idx)
    return xs[idx[:n], :]

try:
    rss = unpickle(os.path.join(base_dir, '%s_rss.npy' % ident))
    rs = unpickle(os.path.join(base_dir, '%s_rs.npy' % ident))
    cs = unpickle(os.path.join(base_dir, '%s_cs.npy' % ident))
except:
    yss = [sample(xs, n) for xs in xss]
    m = stats.HDPMixtureModel(nclusts=nclusts, burnin=burnin, niter=niter)
    m.ident = True
    rss = m.fit(yss, verbose=100)
    pickle(os.path.join(base_dir, '%s_rss.npy' % ident), rss)
    print "rss saved"
    rs = [r.average() for r in rss]
    pickle(os.path.join(base_dir, '%s_rs.npy' % ident), rs)
    print "rs saved"
    cs = [r.make_modal() for r in rs]
    pickle(os.path.join(base_dir, '%s_cs.npy' % ident), cs)
    print "cs saved"
```
